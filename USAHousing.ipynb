```python
df['LotFrontage'].fillna(method='ffill',inplace=True)
df['LotFrontage']
```

```python
df['MSZoning'].fillna(method='bfill',inplace=True)
```

```python
df['SaleType'].fillna(method='backfill',inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X = df['LotFrontage']
y = df['MoSold']
```

```python
import numpy as np

X_shape = np.reshape(X,(-1,1))
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X_shape,y,test_size=0.2,random_state=42)

model.fit(X_train,y_train)
```

```python
predictt=model.predict(X_test)
predictt
```
